###Carregando os dados

```python
import pandas as pd
url = 'https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv'
dados = pd.read_csv(url)
dados.head()
```

###Desafio 1
###### - Realizar a média da metragem para cada um dos bairros. 

```python
#nomes dos bairros
nomes_bairros = dados['Bairro'].values 
dict_media = {}  #criando uma dicionário vazio

for bairro in nomes_bairros:
  media_metragem = dados.loc[dados['Bairro'] == bairro, 'Metragem'].mean() #calcula a média da metragem de cada bairro
  dict_media[bairro] = media_metragem # Adiciona a média ao dicionário
```

###Desafio 3
Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, assim como tentar colocar alguma conclusão.

```python
bairros_mais_imoveis = dados['Bairro'].value_counts().head(10)
```

```python
#Os dez bairros com mais moradias 
bairros_mais_imoveis.plot(kind="bar") 
```

>Os 10 bairros com o maior percentual de moradias disponíveis, são os vistos no grafico de barras acima.

```python
#contando as moradias que restaram
bairros_moradias_restantes = dados['Bairro'].value_counts().sum() - bairros_mais_imoveis.sum()
bairros_mais_imoveis['Outros bairros'] = bairros_moradias_restantes

#criando um gráfico
bairros_mais_imoveis.plot(kind="pie", autopct='%1.1f%%', fontsize=8, figsize=(14, 14))
```

>Acima podemos ver que com o gráfico **pie** podemos  ver o percentual de moradias disponíveis dos bairros citados anteriormente, em relação aos demais bairros.

###Desafio 5
Descobrir quais são os bairros que não tem nome de rua. 

```python
bairros_sem_rua = []

for bairro in dados['Bairro'].values:
  nao_nulos = dados.loc[dados['Bairro'] == bairro, 'Rua'].notnull().sum() # contando os valores non null.
  if nao_nulos == 0: 
    bairros_sem_rua.append(bairro)

bairros_sem_rua
```
