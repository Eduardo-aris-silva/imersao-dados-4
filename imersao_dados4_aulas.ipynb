###Treinando o modelo utilizando um algoritimo de **Regressão Linear** com o **Scikit-learn**

> "Em estatística ou econometria, [*regressão linear*](https://pt.wikipedia.org/wiki/Regress%C3%A3o_linear) é uma equação para se estimar a condicional de uma variável y, dados os valores de algumas outras variáveis x. A regressão, em geral, tem como objectivo tratar de um valor que não se consegue estimar inicialmente..." 
Para mais informações [clique aqui](https://www.alura.com.br/artigos/desvendando-a-regressao-linear).



####1º modelo

```python
#importando o scikit-learn
from sklearn.linear_model import LinearRegression #importando o algoritimo de regressão linear
from sklearn.model_selection import train_test_split 
#Definindo as variáveis que serão usadas
X = dados_vendas_censo[["Metragem"]]
Y = dados_vendas_censo["Valor_anuncio"]
#Dividindo os dados em dados de treino e dados de teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr = LinearRegression()
```

```python
X_train
```

```python
#treinando o modelo
lr.fit(X_train, Y_train)
```

```python
#avaliando a precisão do modelo
Y_predict_test = lr.predict(X_test)
```

```python
Y_test
```

```python
#importando uma métrica para avaliar a acurácia do modelo
from sklearn.metrics import mean_absolute_error, r2_score

mean_absolute_error(Y_test, Y_predict_test)
```

####2º modelo

```python
#Definindo as entradas(X) e a saída(Y)
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V001','V007', 'V009']]
Y = dados_vendas_censo["Valor_anuncio"]
#Dividindo os dados (treino e teste)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr.fit(X_train, Y_train)

Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
#Métricas de teste do modelo
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```
